```python
def remove_cycles(actions):
    def complementary(action_1, action_2):
        return action_1.strip("'") == action_2.strip("'") and (
                (action_1[-1] == "'" and action_2[-1] != "'")
                or (action_1[-1] != "'" and action_2[-1] == "'")
            )

    if (
        len(actions) > 1 and
        # opposite moves
        complementary(actions[-1], actions[-2])
    ):
        actions = actions[:-2]
    if (
        len(actions) > 3 and
        # 360 degree rotation
        len(set(actions[-4:]))==1
    ):
        actions = actions[:-4]

    return actions
```
